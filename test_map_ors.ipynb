```python
import requests
import os
import json
from pprint import pprint
from dotenv import load_dotenv

# Load API key from .env file
load_dotenv()
ors_key = os.getenv("OPENROUTESERVICE_API_KEY")

# Correct coordinate format & added "units"
body = {
    "coordinates": [[-79.3957, 43.6629], [-79.3806, 43.6544]],
    "units": "m"
}

headers = {
    'Accept': 'application/json, application/geo+json, application/gpx+xml, img/png; charset=utf-8',
    'Authorization': ors_key,
    'Content-Type': 'application/json; charset=utf-8'
}

# Make API request
call = requests.post('https://api.openrouteservice.org/v2/directions/driving-car', json=body, headers=headers)

# Print status code
print(call.status_code, call.reason)

# Print formatted JSON response
if call.status_code == 200:
    response_data = call.json()
    pprint(response_data)  # Pretty-print JSON
else:
    print("Error:", call.text)
```
